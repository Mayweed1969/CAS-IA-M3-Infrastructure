```sql
-- Calcule le total des remises par produit et identifie celui avec la remise totale la plus élevée
WITH total_remises_par_produit AS (
  SELECT
    ProductID,
    SUM(UnitPriceDiscount) AS total_remise
  FROM amu_adventureworks.saleslt.salesorderdetail
  GROUP BY ProductID
)
SELECT
  ProductID,
  total_remise
FROM total_remises_par_produit
ORDER BY total_remise DESC
LIMIT 1
```
