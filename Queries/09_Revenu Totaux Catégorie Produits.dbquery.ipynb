```sql
SELECT 
    pc.Name AS ProductCategory,
    SUM(sod.LineTotal) AS TotalRevenue
FROM amu_adventureworks.saleslt.salesorderheader soh
JOIN amu_adventureworks.saleslt.salesorderdetail sod
    ON soh.SalesOrderID = sod.SalesOrderID
JOIN amu_adventureworks.saleslt.product p
    ON sod.ProductID = p.ProductID
JOIN amu_adventureworks.saleslt.productcategory pc
    ON p.ProductCategoryID = pc.ProductCategoryID
GROUP BY pc.Name
ORDER BY TotalRevenue DESC
```
